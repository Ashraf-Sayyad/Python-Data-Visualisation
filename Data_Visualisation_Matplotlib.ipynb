```python
x = np.random.rand(50)
y= np.random.rand(50)

colours = np.random.rand(50)
sizes = 1000 *np.random.rand(50)
plt.scatter(x, y, c = colours, s= sizes, alpha= .5) # alpha = intensity of colur
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.figure(figsize=(6,3))
plt.plot(x, y, '--b')
plt.xlabel("this is my x label data")
plt.ylabel("this is my y label data")
plt.title("this is my title")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y= np.random.rand(50)

colours = np.random.rand(50)
sizes = 1000 *np.random.rand(50)
plt.scatter(x, y, c = '#A64D13', s= sizes, alpha= 0.8,
            ) # alpha = intensity of colur
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
x = np.random.rand(50)
y= np.random.rand(50)

colours = np.random.rand(50)
sizes = 1000 *np.random.rand(50)
plt.scatter(x, y, c = '#A64D13', s= sizes, alpha= 0.8, marker='x'
            ) # alpha = intensity of colur
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```
